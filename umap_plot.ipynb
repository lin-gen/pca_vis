```python
da = pd.read_csv("../consensus_wmb/integrated.cellmeta.csv.gz",index_col=0)
```

```python
levels, categories = pd.factorize(da['annotation']) 
palette = dict(zip(categories,cc.glasbey_bw[:len(categories)]))
fig, ax = plt.subplots()
ax.scatter(da["UMAP1"],da["UMAP2"], c=pd.Series(palette.values())[levels],alpha = 0.1,s=0.5)
for label in categories:
    ax.scatter([], [], c=np.array(palette[label]).reshape(1,-1), label=label)
ax.legend(
    frameon=False,
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    ncol=2,
    fontsize=10
            )
fig.savefig("umap_annotation.png")
```
